```python
#1
#a.
import numpy as np
import pandas as pd

df1 = pd.read_csv("Desktop/hwk06-01.csv")
```

```python
#b.
columns = ['department','status', 'age', 'salary']
df2 = pd.DataFrame(columns=columns)
i = 0
for index, row in df1.iterrows():
    for r in range(row[4]):
        df2.loc[i] = [row[0],row[1],row[2],row[3]]
        i += 1
        
print(df2)
```

```python
#c.
df2_ = df2.copy()

encode = {
  "sales": 0,
  "systems": 1,
  "marketing": 2,
  "secretary": 3,
  "senior": 4,
  "junior": 5,
  "31..35": 6,
  "26..30": 7,
  "21..25": 8,
  "41..45": 9,
  "36..40": 10,
  "46..50": 11,
  "46k..50k": 12,
  "26k..30k": 13,
  "31k..35k": 14,
  "66k..70k": 15,
  "41k..45k": 16,
  "36k..40k": 17,
}

#d.
for index, row in df2_.iterrows():
    row[0] = encode.get(row[0])
    row[1] = encode.get(row[1])
    row[2] = encode.get(row[2])
    row[3] = encode.get(row[3])

print(df2_)
```

```python
#2
m1 = df2_.drop('status', axis=1)
m2 = df2_['status']
m1 = m1.astype(int)
m2 = m2.astype(int)

from sklearn.model_selection import train_test_split

m1Train, m1Test, m2Train, m2Test = train_test_split(m1, m2)

from sklearn import tree

model = tree.DecisionTreeClassifier()

model.fit(m1Train, m2Train)

#Need to install graphviz package 
import graphviz 
from graphviz import Source
dot_data = tree.export_graphviz(model, out_file=None, feature_names=m1Train.columns,class_names=["senior","junior"])
graph = graphviz.Source(dot_data) 
graph.render("gTree",view = True)

#t =< department : systems, status :?, age : 28, salary : 50K > 
#Predicts that status is = junior

columns = ['department','age', 'salary']
test = pd.DataFrame(columns=columns)
test.loc[0] = [encode.get("systems"),encode.get("26..30"),encode.get("66k..70k")]
print("t =< department : systems, status :?, age : 28, salary : 50K > ")
print("ENCODED:")
print(test)
m2Predict = model.predict(test)
print("Decision Tree Prediction:")
if(m2Predict == 5):
    print("*Junior*")
else:
    print("*Senior*")
```

```python
#3
#a.
columns = ['department','status', 'age', 'salary']
df3 = pd.DataFrame(columns=columns)
i = 0
for index, row in df1.iterrows():
    for r in range(row[4]):
        df3.loc[i] = [row[0],row[1],row[2],row[3]]
        i += 1
        
print(df3)
df3_ = df3.copy()

#b
#c

import random
print 
encodeRand = {
  "sales": 0,
  "systems": 1,
  "marketing": 2,
  "secretary": 3,
  "senior": 4,
  "junior": 5,
  "31..35": 10,
  "26..30": 11,
  "21..25": 12,
  "41..45": 13,
  "36..40": 14,
  "46..50": 15,
  "46k..50k": 15,
  "26k..30k": 11,
  "31k..35k": 10,
  "66k..70k": 16,
  "41k..45k": 13,
  "36k..40k": 14,
}

def getRandVal(val):
    if(val == 10):
        return random.randint(30, 35)
    elif(val == 11):
        return random.randint(25, 30)
    elif(val == 12):
        return random.randint(20, 25)
    elif(val == 13):
        return random.randint(40, 45)
    elif(val == 14):
        return random.randint(35, 40)
    elif(val == 15):
        return random.randint(45, 50)
    elif(val == 16):
        return random.randint(65, 70)
    else:
        return 0

for index, row in df3_.iterrows():
    row[0] = encodeRand.get(row[0])
    row[1] = encodeRand.get(row[1])
    row[2] = getRandVal(encodeRand.get(row[2]))
    row[3] = getRandVal(encodeRand.get(row[3]))
#d
print(df3_)
```

```python
#4

m1 = df3_.drop('status', axis=1)
m2 = df3_['status']
m1 = m1.astype(int)
m2 = m2.astype(int)

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

m1Train, m1Test, m2Train, m2Test = train_test_split(m1, m2, random_state=1)

gNB = GaussianNB()

gNB.fit(m1Train, m2Train)

#t =< department : systems, status :?, age : 28, salary : 50K > 
#Predicts that status is = junior

columns = ['department','age', 'salary']
test = pd.DataFrame(columns=columns)
test.loc[0] = [encodeRand.get("systems"),28,getRandVal(encodeRand.get("66k..70k"))]
print("t =< department : systems, status :?, age : 28, salary : 50K > ")
print("ENCODED:")
print(test)
m2Predict = gNB.predict(test)
print("Gaussian Naive Bayes Prediction:")
if(m2Predict == 5):
    print("*Junior*")
else:
    print("*Senior*")
```
